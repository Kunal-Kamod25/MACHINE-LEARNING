```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
data.iloc[:,1:-1].values
X = data.values  
K = 3
```

```python
n= data.shape[0]
m = data.shape[1]
k = n*m
n
```

```python
# Use the full iris dataset for K-means clustering
X = data.values
# Show first two samples as example
a = X[0]
b = X[1]
c = X[2]
a, b, c
```

```python
def calc_distance(x1,x2):
    # handle 1D numeric vectors: sum over all elements
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
# compute distance between the two sampled rows
calc_distance(X[0], X[1])
```

```python
# choose two random indices from X safely
np.random.choice(len(X), size=2, replace=False)
```

```python
# pick k unique random indices for initial centroids
random_index = np.random.choice(len(X),k)
centroids = X[random_index]
```

```python
distance0 = []
distance1 = []
distance2 = []

for i in range(n):
    dist = calc_distance(X[i], centroids[0])
    distance0.append(dist)
    
for i in range(n):
    dist = calc_distance(X[i], centroids[1])
    distance1.append(dist)

for i in range(n):
    dist = calc_distance(X[i], centroids[2])
    distance2.append(dist)
```

```python
labels = []
for i in range(n):
    labels.append(np.argmin([distance0[i], distance1[i], distance2[i]]))
```
